```python
def recipe_features(row):
    description_length = len(str(row['description'])) / 6321
    minutes = row['minutes']
    n_steps = row['n_steps']
    n_ingredients = row['n_ingredients']
    return np.array([1, description_length, minutes, n_steps, n_ingredients])
```

```python
recipe_feature_dict = {row['id']: recipe_features(row) for _, row in raw_recipes.iterrows()}
```

```python
def user_features(row, userRatings):
    avg_rating = np.mean(userRatings[row['ratings']])
    n_items = row['n_items']
    n_ratings = row['n_ratings']
    return np.array([1, avg_rating, n_items, n_ratings])
```

```python
user_feature_dict = {row['u']: user_features(row, userRatings) for _, row in users_df.iterrows()}
```

## Eucledian dist

```python
def similarity(user_id, recipe_id):
    user_feature = user_feature_dict.get(user_id, np.zeros(4))
    recipe_feature = recipe_feature_dict.get(recipe_id, np.zeros(5))

    if len(user_feature) < len(recipe_feature):
        user_feature = np.pad(user_feature, (0, len(recipe_feature) - len(user_feature)), 'constant')
    elif len(recipe_feature) < len(user_feature):
        recipe_feature = np.pad(recipe_feature, (0, len(user_feature) - len(recipe_feature)), 'constant')
    
    return euclidean(user_feature, recipe_feature)
```
